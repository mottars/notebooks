# Exemple: using SciPy Python for Reliability Analysis
    

#### Importing packages

```python
import scipy.stats as stat
import numpy as np
import matplotlib.pyplot as plt
```

**Define Random Variables**

```python
mean = 4
st_d = 1
VA = stat.norm(loc=mean, scale=st_d)
```

**Plotar pdf**

```python
x=np.linspace(0,8,100)
plt.plot(x,VA.pdf(x))
```

**Obtaining 90% Confidence Interval** 

[P(5%), P(95%)]

```python
x_min = VA.ppf(0.05)
x_max = VA.ppf(0.95)
print([x_min,x_max])
```

checking...

```python
print([VA.cdf(x_min),VA.cdf(x_max)])
```

**Sample Generation**

```python
N = 100000
xs = VA.rvs(size=N)
print([np.mean(xs),np.std(xs)])
```

**Histogram**

```python
a = plt.hist(xs,bins=30)
```

**Function of Random Variable**

```python
ys = 5/(xs**2+3)
a=plt.hist(ys,bins=30)
```

#### Failure Function: g = 1 - y

```python
g = 1 - ys
plt.plot(xs,g,'.')
```

#### Probability (Relative Frequency) of Failure: P(g<0)

```python
nfail = np.sum(g<0)
PF = nfail/N
print('number of failures = %d'%nfail)
print('Probability of Failures, PF = %f'%PF)
```
